```python
no_funky_business = {
    'height': 141 * 6,
    'width': 600,
    'p_margin': (0.1, 0.1, 0.1, 0.15),
    'p_texture_noise': (0, 0),
    'p_line_spacing': (60, 0),
    'p_line_angle': (0, 0),
    'p_line_offset': (0, 0),
    'p_line_thickness': (0.004, 0),
    'p_line_intensity': (0, 0),
    'p_line_dropout': 0.0,
    'p_line_breaker': (0, 0, 0),
    'p_vignette_blur': 0,
}
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
paper, count = generate_paper(text=".           LINE DETECTION                        .               USING                        .         HOUGH TRANSFORM                     ", p_line_dropout=1.0)
ax.set_title("Random Text")
ax.imshow(paper, cmap="gray", vmin=0, vmax=255)
plt.show()
```

## Simple Paper

None of the complications are present.

```python
fig, axs = plt.subplots(1, 2, figsize=(6.6, 4))
paper, count = generate_paper(**no_funky_business)
axs[0].set_title("Random Text")
axs[0].imshow(paper, cmap="gray", vmin=0, vmax=255)
paper, count = generate_paper(text="", **no_funky_business)
axs[1].set_title("Plain Paper")
axs[1].imshow(paper, cmap="gray", vmin=0, vmax=255)
plt.suptitle("Simple Paper")
plt.show()
```

## Effect of Paper Texture

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
for index, p_texture_noise in enumerate([(0, 0), (0, 10), (0, 30)]):
    paper, count = generate_paper(text="", **{**no_funky_business, 'p_texture_noise': p_texture_noise})
    axs[index].imshow(paper, cmap="gray", vmin=0, vmax=255)
plt.suptitle("Effect of Paper Texture")
plt.show()
```
